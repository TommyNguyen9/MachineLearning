```python
X = df.drop('Outcome', axis = 'columns')
y = df.Outcome
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled[:3]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, stratify = y, random_state = 17)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score

scores = cross_val_score(DecisionTreeClassifier(), X, y, cv = 7)
scores
```

```python
scores.mean()
```

```python
from sklearn.ensemble import BaggingClassifier

bag_model = BaggingClassifier(
    estimator = DecisionTreeClassifier(),
    n_estimators = 125,
    max_samples = 0.9,
    oob_score = True, # Out of bag.
    random_state = 0

)

bag_model.fit(X_train, y_train)
bag_model.oob_score_
```

```python
bag_model.score(X_test, y_test)
```

```python
bag_model = BaggingClassifier(
    estimator = DecisionTreeClassifier(),
    n_estimators = 125,
    max_samples = 0.9,
    oob_score = True, # Out of bag.
    random_state = 0

)

scores = cross_val_score(bag_model, X, y, cv = 8)
scores.mean()
```

```python
from sklearn.ensemble import RandomForestClassifier

scores = cross_val_score(RandomForestClassifier(), X, y, cv = 6)
scores.mean()
```
