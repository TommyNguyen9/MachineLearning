```python
import pandas as pd

df = pd.read_csv("heart.csv")
df 
```

```python
import numpy as np
from scipy.stats import zscore
import matplotlib.pyplot as plt
```

```python
print(df.columns)
```

```python
df = df[df['Cholesterol'] != 0]
df = df[df['RestingBP'] != 0]
```

```python
# Remove outliers:

cols = ['MaxHR', 'Cholesterol', 'RestingBP', 'Oldpeak']
df['Z'] = df[cols].apply(zscore).abs().max(axis = 1)

df_clean = df[df['Z'] < 3]
```

```python
df_dummies = pd.get_dummies(df, 
                            columns = ['Sex', 'ChestPainType', 'RestingECG', 'ExerciseAngina', 'ST_Slope'],
                            drop_first = True)
```

```python
print(df_dummies.columns)
```

```python
df_dummies.keys()
```

```python
df_dummies
```

```python
df_dummies.describe()
```

```python
X = df_dummies.drop('HeartDisease', axis = 1)
y = df_dummies['HeartDisease']
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size = 0.38, random_state = 94)
```

```python
from sklearn.svm import SVC
model = SVC(kernel = 'linear')
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
model.fit(X_scaled, y)
```

```python
model.score(X_scaled, y)
```

```python
from sklearn.ensemble import BaggingClassifier

bag_model = BaggingClassifier(
    estimator = SVC(),
    n_estimators = 140,
    max_samples = 0.95,
    oob_score = True,
    random_state = 0
)

bag_model.fit(X_train, y_train)
bag_model.oob_score_
```

```python
bag_model.score(X_test, y_test)
```

```python
# Decision tree:

from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score

scores = cross_val_score(DecisionTreeClassifier(), X, y, cv = 9)
scores
```

```python
scores.mean()
```

```python
bag_model = BaggingClassifier(
    estimator = DecisionTreeClassifier(),
    n_estimators = 160,
    max_samples = 0.84,
    oob_score = True,
    random_state = 0
)

scores = cross_val_score(bag_model, X, y, cv = 9)
scores.mean()
```

```python
scores
```

<p style = 'font-size: 14px;">

### It is useful to use bagging during SVC and Decision Tree models because this will increase the predicted score. 

### Bagging helps reduce variance within a noisy dataset i.e: high variance and low bias models eg: KNNs, fully grown decision trees.
