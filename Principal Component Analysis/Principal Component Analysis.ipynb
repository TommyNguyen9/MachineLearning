```python
df = pd.DataFrame(digits.data, columns = digits.feature_names)
df.head()
```

```python
df.describe()
```

```python
X = df
y = digits.target
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test =  train_test_split(X_scaled, y, test_size = 0.35, random_state = 20)
```

```python
from sklearn.linear_model import LogisticRegression

ls = LogisticRegression()
ls.fit(X_train, y_train)
ls.score(X_test, y_test)
```

```python
X.shape
```

```python
from sklearn.decomposition import PCA

pca = PCA(0.83)

X_pca = pca.fit_transform(X)
X_pca.shape
```

```python
pca.explained_variance_ratio_
```

```python
pca.n_components_
```

```python
X_train_pca, X_test_pca, y_train, y_test = train_test_split(X_pca, y, test_size = 0.35, random_state = 20)
```

```python
ls = LogisticRegression(max_iter = 1000)
ls.fit(X_train_pca, y_train)
ls.score(X_test_pca, y_test)
```

```python
pca = PCA(n_components = 3)
X_pca = pca.fit_transform(X)
X_pca.shape
```

```python
X_pca
```

```python
pca.explained_variance_ratio_
```

```python
X_train_pca, X_test_pca, y_train, y_test = train_test_split(X_pca, y, test_size = 0.35, random_state = 20)

ls = LogisticRegression(max_iter = 1200)
ls.fit(X_train_pca, y_train)
ls.score(X_test_pca, y_test)

```
