```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size = 0.3)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
import numpy as np
```

```python
lr = LogisticRegression()
lr.fit(X_train, y_train)
lr.score(X_test, y_test)
```

```python
svm = SVC()
svm.fit(X_train, y_train)
svm.score(X_test, y_test)
```

```python
rf = RandomForestClassifier(n_estimators = 40)
rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```

```python
from sklearn.model_selection import KFold
kf = KFold(n_splits = 9)
kf
```

```python
for train_index, test_index in kf.split([1, 2, 3, 4, 5, 6, 7, 8, 9]):
    print(train_index, test_index)
```

```python
def get_score(model, X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)
    return model.score(X_test, y_test)
```

```python
get_score(LogisticRegression(), X_train, X_test, y_train, y_test)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import StratifiedKFold

folds = StratifiedKFold(n_splits = 7)
```

```python
import numpy as np
np.set_printoptions(precision = 2)
```

```python
scores_l = []
scores_svm = []
scores_rf = []

for train_index, test_index in kf.split(iris.data):
    X_train, X_test, y_train, y_test = (
        iris.data[train_index], iris.data[test_index],
        iris.target[train_index], iris.target[test_index]
    )

    scores_l.append(get_score(LogisticRegression(), X_train, X_test, y_train, y_test))
    scores_svm.append(get_score(SVC(), X_train, X_test, y_train, y_test))
    scores_rf.append(get_score(RandomForestClassifier(n_estimators = 80), X_train, X_test, y_train, y_test))
```

```python
cross_val_score(SVC(), iris.data, iris.target)
```

```python
np.average(scores_svm)
```

```python
cross_val_score(LogisticRegression(), iris.data, iris.target)
```

```python
np.average(scores_l)
```

```python
cross_val_score(RandomForestClassifier(n_estimators = 80), iris.data, iris.target)
```

```python
np.average(scores_rf)
```
